```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

# Метод трапеций

```python
def trapezoid_method(a, b, N, f):
    xi = a
    xj = a + (b - a) / N
    h = (b - a) / N
    I = 0
    for i in range(N):
        I += (f(xi) + f(xj)) * h / 2
        xi += h
        xj += h
    return I
```

# Метод Симпсона

```python
def simpsons_method(a, b, N, f):
    xi = a
    xj = a + (b - a) / N
    h = (b - a) / N
    I = 0
    for i in range(N):
        I += (f(xi) + 4 * f((xj + xi) / 2) + f(xj)) * h / 6
        xi += h
        xj += h
    return I
```

# Первый интеграл

```python
N = np.array([2 ** i for i in range(2, 6)])
true_solution = [1.5708] * N.shape[0]
trapezoidal_solutions = np.zeros_like(N, dtype=float)
simpsons_solutions = np.zeros_like(N, dtype=float)
j = 0
for i in N:
    trapezoidal_solutions[j] = trapezoid_method(-1, 1, i, lambda x: 1 / (1 + x ** 2))
    simpsons_solutions[j] = simpsons_method(-1, 1, i, lambda x: 1 / (1 + x ** 2))
    j += 1
```

```python
plt.scatter(N, true_solution, label='true')
plt.scatter(N, trapezoidal_solutions, label='trapezoid', marker='+')
plt.scatter(N, simpsons_solutions, label='simpsons', marker='x')
plt.grid()
plt.xticks(N)
plt.legend()
```

# Второй интеграл

```python
N = np.array([2 ** i for i in range(2, 8)])
true_solution = [1.2958] * N.shape[0]
trapezoidal_solutions = np.zeros_like(N, dtype=float)
simpsons_solutions = np.zeros_like(N, dtype=float)
j = 0
for i in N:
    trapezoidal_solutions[j] = trapezoid_method(0, 1, i, lambda x: np.power(x, 1 / 3) * np.exp(np.sin(x)))
    simpsons_solutions[j] = simpsons_method(0, 1, i, lambda x: np.power(x, 1 / 3) * np.exp(np.sin(x)))
    j += 1
```

```python
plt.scatter(N, true_solution, label='true')
plt.scatter(N, trapezoidal_solutions, label='trapezoid', marker='+')
plt.scatter(N, simpsons_solutions, label='simpsons', marker='x')
plt.grid()
plt.xticks(N)
plt.legend()
```
