```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

# Метод трапеций

```python
def trapezoid_method(a, b, N, f):
    xi = a
    xj = a + (b - a) / N
    h = (b - a) / N
    I = 0
    for i in range(N):
        I += (f(xi) + f(xj)) * h / 2
        xi += h
        xj += h
    return I
```

# Метод Симпсона

```python
def simpsons_method(a, b, N, f):
    xi = a
    xj = a + (b - a) / N
    h = (b - a) / N
    I = 0
    for i in range(N):
        I += (f(xi) + 4 * f((xj + xi) / 2) + f(xj)) * h / 6
        xi += h
        xj += h
    return I
```

# Первый интеграл

```python
N = np.array([2 ** i for i in range(11)])
true_solution = np.array([np.pi / 2] * N.shape[0])
trapezoidal_solutions = np.zeros_like(N, dtype=np.double)
simpsons_solutions = np.zeros_like(N, dtype=np.double)
j = 0
for i in N:
    trapezoidal_solutions[j] = trapezoid_method(-1, 1, i, lambda x: 1 / (1 + x ** 2))
    simpsons_solutions[j] = simpsons_method(-1, 1, i, lambda x: 1 / (1 + x ** 2))
    j += 1
```

```python
plt.plot(N, np.abs(trapezoidal_solutions-true_solution), label='trapezoid', marker='*')
plt.plot(N, np.abs(simpsons_solutions-true_solution), label='simpsons', marker='*')
n2 = np.array([1/(i**2) for i in np.linspace(0.1, 200, num=500)])
n4 = np.array([1/(i**4) for i in np.linspace(0.1, 200, num=500)])
plt.plot(n2)
plt.plot(n4, color='k')
plt.grid()
plt.xticks(N)
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```

# Второй интеграл

```python
N = np.array([2 ** i for i in range(2, 15)])
true_solution = np.array([1.295874008731708] * N.shape[0])
trapezoidal_solutions = np.zeros_like(N, dtype=float)
simpsons_solutions = np.zeros_like(N, dtype=float)
j = 0
for i in N:
    trapezoidal_solutions[j] = trapezoid_method(0, 1, i, lambda x: np.power(x, 1 / 3) * np.exp(np.sin(x)))
    simpsons_solutions[j] = simpsons_method(0, 1, i, lambda x: np.power(x, 1 / 3) * np.exp(np.sin(x)))
    j += 1
```

```python
plt.plot(N, np.abs(true_solution - trapezoidal_solutions), label='trapezoid', marker='*')
plt.plot(N, np.abs(true_solution-simpsons_solutions), label='simpsons', marker='*')
plt.grid()
plt.xticks(N)
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
